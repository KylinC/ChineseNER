## torch LSTM

```
def prepare_sequence(seq,to_ix):
    idxs=[to_ix[w] for w in seq]
    return torch.tensor(idxs,dtype=torch.long)
```

```
tag_to_ix={"nn":0,"nr":1,"nt":2}
EMBEDDING_DIM=20
HIDDEN_DIM=20
```

```
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
 
torch.manual_seed(1)
```

```
cut_all = []
for sample in train:
  sentence = sample[0][0]
  words = pseg.cut(sentence)
  words_seq = []
  for word, flag in words:
    words_seq.append(word)
  cut_all.append(words_seq)
print(cut_all)
```

```
print(len(cut_all)==len(all_tag_list))
```

```
a = all_tag_list[0]
print(list(a.keys()))
```

```
aim_all = []
for idx in range(len(cut_all)):
  aim_tmp = []
  split_sentence = cut_all[idx]
  keys_list = all_tag_list[idx].keys()
  for item in split_sentence:
    if item in keys_list:
      aim_tmp.append(all_tag_list[idx][item])
    else:
      aim_tmp.append('nn')
  aim_all.append(aim_tmp)
print(aim_all)
```

```
with torch.no_grad():
    test_matrix = prepare_sequence(cut_all[0],word_to_ix)
```

```
class LSTMTagger(nn.Module):
    def __init__(self,embedding_dim,hidden_dim,vocab_size,tagset_size):
        super(LSTMTagger,self).__init__()
        self.hidden_dim=hidden_dim
        self.word_embeddings=nn.Embedding(vocab_size,embedding_dim)
        self.lstm=nn.LSTM(embedding_dim,hidden_dim)
        self.hidden2tag=nn.Linear(hidden_dim,tagset_size)
        self.hidden=self.init_hidden()
    def init_hidden(self):
        #the axes semantics are (num_layers,minibatch_size,hidden_size)
        return (torch.zeros(1,1,self.hidden_dim),
                (torch.zeros(1,1,self.hidden_dim)))
    def forward(self, sentence):
        embeds=self.word_embeddings(sentence)
        lstm_out,self.hidden=self.lstm(embeds.view(len(sentence),1,-1),self.hidden)
        tag_space=self.hidden2tag(lstm_out.view(len(sentence),-1))
        tag_scores=F.log_softmax(tag_space,dim=1)
        return tag_scores
```

```
model=LSTMTagger(EMBEDDING_DIM,HIDDEN_DIM,len(word_to_ix),len(tag_to_ix))
loss_function=nn.NLLLoss()
optimizer=optim.SGD(model.parameters(),lr=0.1)
```

```
for epoch in range(500):
    for idx in range(len(cut_all)):
      sentence = cut_all[idx]
      tags = aim_all[idx]
      model.zero_grad()
      model.hidden=model.init_hidden()
      sentence_in=prepare_sequence(sentence,word_to_ix)
      targets=prepare_sequence(tags,tag_to_ix)
      tag_scores=model(sentence_in)
      loss=loss_function(tag_scores,targets)
      loss.backward()
      optimizer.step()
    print(epoch)
```

```
with torch.no_grad():
    test_number = 14


    inputs=prepare_sequence(cut_all[test_number],word_to_ix)
    tag_scores=model(inputs)
    
    employ = []
    employee = []
    for idx in range(len(tag_scores)):
      if(tag_scores[idx][1]>tag_scores[idx][0] and tag_scores[idx][1]>tag_scores[idx][2]):
        employee.append(cut_all[test_number][idx])
      if(tag_scores[idx][2]>tag_scores[idx][0] and tag_scores[idx][2]>tag_scores[idx][1]):
        employ.append(cut_all[test_number][idx])

print(employee)
print(employ)
```
