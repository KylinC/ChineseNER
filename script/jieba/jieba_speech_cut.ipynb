```
all_list = []
for non_split_item in split_dict1:
  tmp_sentence_list = []
  tmp_tag_list = []
  # split the sentence for detaching tag and content
  tmp_split_list = [x for x in non_split_item.split('\n') if x]
  split_length = len(tmp_split_list)
  
  if (split_length == 1):
    tmp_sentence_list.append(tmp_split_list[0])
  else:
    tmp_sentence_list.append(tmp_split_list[0])
    tmp_tag_list += tmp_split_list[1:]
    
  # add to all_list
  all_list.append([tmp_sentence_list,tmp_tag_list])

```
