```
# random vectors
x=torch.randn((3,4,5))
print(x)
```

```
# simple algebraic operation
x=torch.tensor([1,2,3])
y=torch.tensor([4,5,6])
z=x+y
print(z)
```

```
#By default, concatenates rows
x_1=torch.randn(2,5)
y_1=torch.randn(3,5)
z_1=torch.cat([x_1,y_1],0)
print(x_1)
print(y_1)
print(z_1)
```

```
#concatenate columns
x_2=torch.randn(2,3)
y_2=torch.randn(2,2)
z_2=torch.cat([x_2,y_2],1)
print(x_2)
print(y_2)
print(z_2)
```

```
x=torch.randn(2,3,4)
print(x)
print(x.view(2,12))
print(x.view(2,-1)) # default
print(x.view(3,-1)) # default
```

```
x=torch.tensor([1.,2,3],requires_grad=True)
y=torch.tensor([4.,5,6],requires_grad=True)
print(x.grad_fn)
z=x+2*y
print(z)
print(z.grad_fn)
```

```
# backward computation
s=z.sum()
print(s)
print(s.grad_fn)
```

```
s.backward()
print(z.grad_fn)
print(x.grad)
print(y.grad)
print(z.grad)
```

```
x=torch.randn(2,2)
y=torch.randn(2,2)
print(x.requires_grad,y.requires_grad)
z=x+y
print(z.requires_grad)
print(z.grad_fn)
x=x.requires_grad_()
print(x.requires_grad)
y=y.requires_grad_()
z=x+y
print(z.grad_fn)
print(z.requires_grad)
new_z=z.detach()
print(new_z.grad_fn)
```
